# Gradient Descent

```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = symbols('x')
#exp1 = exp(x) - ln(x)
exp1 = x**3 - 2*x -5
x0 = 0
tol = 0.001
ss = 0.01
gradient = diff(exp1,x)
slope = gradient.subs(x,x0)
x1 = 8
list_x1 = [2]
list_y1 = [exp1.subs(x,2).evalf()]

while abs(x1 - x0) > tol:
    

    x1 = x0
    x0 = x0 - gradient.subs(x,x0) * ss

    list_x1.append(x0.evalf())
    y0 = exp1.subs(x,list_x1[-1])
    list_y1.append(y0.evalf())
    

print(x0.evalf())
print(list_y1)
print(list_x1)
```

# Newton-Raphson method

```python
from sympy import symbols, exp, log
from sympy.utilities.lambdify import lambdify

x = symbols('x')
#f = exp(x) - log(x)
f = x**3 - 2*x -5    # given expression
f1 = f.diff(x)


f = lambdify(x, f)
f1 = lambdify(x, f1)

x = float(input('Enter the initial value x: '))
n = int(input('Enter the desired decimal places: '))
print("Initial value : ", x)
print("decimal places tol : ", n)

lst_x2 = []
lst_y2 = []
i = 1
status1 = True

while status1:
    g = str(x)
    
    try:
        x1 = x - (f(x) / f1(x))
    except ZeroDivisionError:
        print("Error: Division by zero occurred.")
        continue
    
    print("i =", i, "x =", x, "f(x) =", f(x))
    lst_x2.append(x)
    lst_y2.append(f(x))
    m = str(x1)   # cinverting into string to compare decimal places

    if m[0:n+2] == g[0:n+2]:  # to check if same upto the given decimal place
        status1 = False
    else:
        status1 = True
        x = x1
        i += 1

print("Result is", x)
print('x_values',lst_x2)
print('y_values',lst_y2)
```

# Making graph


```python


x = sp.symbols('x')
equation = x**3 - 2*x -5
equation_func = sp.lambdify(x, equation, 'numpy')
x_values = np.linspace(0.01, 2.5, 100)
y_values = equation_func(x_values)

plt.figure(figsize=(10,20))
plt.plot(x_values, y_values, label='x**3 - 2*x -5')
plt.plot(list_x1,list_y1, marker='o',color='red',markeredgecolor='black',linewidth=0.5,markersize='2')
plt.plot(lst_x2,lst_y2, marker='s',color='yellow',markeredgecolor='black',linewidth=0.5,markersize='2')

plt.title('Graph of x**3 - 2*x -5')
plt.xlabel('x')
plt.ylabel('y')

plt.grid(color = 'gray', linestyle = '--')
plt.show()


# It can be seen below that red line of Gradient descent has more number of iteratioin if we want more accuracy
# where as Newton Raphson though has larger jumps but lesser frequency of progressing step or to sayless complexity for such simpler functions
```
